```python
import numpy as np
import cv2 as cv

```

```python
face_classifier = cv.CascadeClassifier('classifier/haarcascade_frontalface_default.xml')
image = cv.imread('face.jpeg')
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
face = face_classifier.detectMultiScale(gray,1.3,5)
```

```python
if face is ():
    print("No face found")
```

```python
for (x,y,w,h) in face:
    cv.rectangle(image,(x,y),(x+w,y+h),(127,0,255),4)
    cv.imshow("Face Detected",image)
    cv.waitKey()
cv.destroyAllWindows()
cv.waitKey(1)
```

```python
import cv2 as cv
import numpy as np
```

```python
face_class = cv.CascadeClassifier('classifier/haarcascade_frontalface_default.xml')
eye_class = cv.CascadeClassifier('classifier/haarcascade_eye.xml')

image = cv.imread('face.jpeg')
gray = cv.cvtColor(image,cv.COLOR_BGR2GRAY)

face = face_class.detectMultiScale(gray,1.3,5)
if face is ():
    print("No face detected")
for (x,y,w,h) in face:
    cv.rectangle(image,(x,y),(y+h,x+w),(127,0,255),2)
    eye_image = image[y:y+h,x:w+x]
    eye = eye_class.detectMultiScale(eye_image)
    if eye is ():
        print("No eye detected")
    for (ex,ey,ew,eh) in eye:
        cv.rectangle(image,(ex,ey),(ex+ew,ey+eh),(127,0,255),2)
        cv.imshow("Face and eye detection",image)
        cv.waitKey(0)
cv.destroyAllWindows()
cv.waitKey(1)
    
```
